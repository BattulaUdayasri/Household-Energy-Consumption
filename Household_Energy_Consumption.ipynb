```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/household_power_consumption.csv",sep=";",na_values='?',low_memory=False)
df.sample(10)
```

```python
df.columns
```

```python
df.columns=['Date', 'Time', 'Global_active_power', 'Global_reactive_power',
       'Voltage', 'Global_intensity', 'Sub_metering_1', 'Sub_metering_2',
       'Sub_metering_3']

```

```python

df['datetime']=pd.to_datetime(df['Date']+' '+df['Time'], format="%d/%m/%Y %H:%M:%S")
df.drop(['Date','Time'],axis=1,inplace=True)
```

```python
df.sample(10)
```

```python
df.set_index('datetime',inplace=True)

```

```python
df.sample(10)
```

```python
df.dtypes
```

```python
df=df.apply(pd.to_numeric,errors='coerce')
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head(100)
```

```python
daily_df=df.resample('D').mean()
```

```python
daily_df.head(10)
```

```python
daily_df['day_of_week']=daily_df.index.dayofweek
daily_df['is_weekend']=daily_df['day_of_week'].apply(lambda x:1 if x>=5 else 0)
```

```python

daily_df.head(20)
```

```python
import os
if not os.path.exists('models'):
  os.makedirs('models')
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import joblib

```

```python
daily_df['target']=daily_df['Global_active_power'].shift(-1)
daily_df.head()
```

```python

daily_df.isnull().sum()
```

```python
daily_df.dropna(inplace=True)
```

```python
daily_df.isnull().sum()
```

```python
X = daily_df[['Global_active_power', 'Voltage', 'is_weekend', 'day_of_week']]
y=daily_df['target']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,shuffle=False,test_size=0.2)
```

```python
model=RandomForestRegressor(n_estimators=100,random_state=42)
model.fit(X_train,y_train)
preds=model.predict(X_test)
```

```python
preds
```

```python
rmse=mean_squared_error(y_test,preds,)
print(f"Forecast Model RMSE: {rmse:3f}")
```

```python
joblib.dump(model,"models/energy_forecast_model.pkl")
```

```python
from sklearn.ensemble import IsolationForest

anomaly_model = IsolationForest(contamination=0.05, random_state=42)
daily_df['anomaly'] = anomaly_model.fit_predict(daily_df[['Global_active_power']])
daily_df['anomaly_label'] = daily_df['anomaly'].apply(lambda x: " Anomaly" if x == -1 else " Normal")

# Save model
joblib.dump(anomaly_model, "models/anomaly_detector.pkl")
```

```python
def generate_recommendations(df):
    recommendations = []

    for idx, row in df.iterrows():
        tips = []

        if row['is_weekend'] == 1 and row['Global_active_power'] > 3.5:
            tips.append(" High weekend usage  consider reducing AC/TV usage.")
        if row['Voltage'] < 230:
            tips.append(" Low voltage  check appliances or wiring.")
        if row['anomaly'] == -1:
            tips.append(" Spike detected  inspect idle/faulty devices.")
        if row['Global_active_power'] > 4.5:
            tips.append(" High usage  use energy-efficient devices.")

        recommendations.append(" | ".join(tips) if tips else " All good!")

    df['recommendations'] = recommendations
    return df
daily_df = generate_recommendations(daily_df)
```

```python
daily_df.to_csv('final_energy_data.csv')
```
